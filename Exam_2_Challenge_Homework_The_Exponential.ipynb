## 1. Using Euler's formula, compute the product: $e^{ix}\cdot e^{iy}$.

What is the real part (that is the term without a factor of $i$)? Remember: $i^2 = -1$

a. $\sin(x)\cos(y) - \cos(x)\sin(y)$

b. $\cos(x)\cos(y) + \sin(x)\sin(y)$

c. $\cos(x)\cos(y) - \sin(x)\sin(y)$

d. $\sin(x)\cos(y) + \cos(x)\sin(y)$

Using Euler's formula, we have:

$e^{ix} \cdot e^{iy} = (\cos(x) + i\sin(x)) \cdot (\cos(y) + i\sin(y))$

To find the real part, we need to multiply the real parts and the imaginary parts separately and then add them together:

$(\cos(x) + i\sin(x)) \cdot (\cos(y) + i\sin(y))$

$= \cos(x) \cdot \cos(y) + \cos(x) \cdot i\sin(y) + i\sin(x) \cdot \cos(y) + i^2\sin(x) \cdot \sin(y)$

Since $i^2 = -1$, we get:

$= \cos(x) \cdot \cos(y) + i\cos(x) \cdot \sin(y) + i\sin(x) \cdot \cos(y) - \sin(x) \cdot \sin(y)$

Now, we can separate the real and imaginary parts:

Real part $= \cos(x) \cdot \cos(y) - \sin(x) \cdot \sin(y)$

The real part is the term without a factor of $i$. So, the correct answer is option c:

$\cos(x) \cdot \cos(y) - \sin(x) \cdot \sin(y)$

## 2. Let $n$ be an integer. Using Euler's formula we have $e^{inx} = \cos(nx) + i\sin(nx)$

On the other hand, we also have: $e^{inx} = (e^{ix})^n = (\cos(x) + i\sin(x))^n$

Putting both of these together, we obtain *de Moivre's* formula:

$\cos(nx) + i\sin(nx) = (\cos(x) + i\sin(x))^n$

Use the latter to find an expression for $\sin(3x)$ in terms of $\sin(x)$ and $\cos(x)$.

Select all that apply:

a. $\sin(3x) = 3\sin(x)\cos^{2}(x) - \sin^{3}(x)$

b. $\sin(3x) = \cos^{3}(x) - 2\sin^{2}(x)\cos(x)$

c. $\sin(3x) = 4\cos^{3}(x) - 3\cos(x)$

d. $\sin(3x) = 3\sin(x) - 4\sin^{3}(x)$

To find an expression for $\sin(3x)$ using de Moivre's formula, we need to expand $(\cos(x) + i\sin(x))^3$ and equate the real part to $\cos(3x)$ and the imaginary part to $\sin(3x)$. Let's proceed with the expansion:

$(\cos(x) + i\sin(x))^3$

$= \cos^3(x) + 3i\cos^2(x)\sin(x) + 3i^2\cos(x)\sin^2(x) + i^3\sin^3(x)$

Remember that $i^2 = -1$ and $i^3 = -i$, so we can simplify:

$= \cos^3(x) + 3i\cos^2(x)\sin(x) - 3i\cos(x)\sin^2(x) - \sin^3(x)$

Now, equating the real part to $\cos(3x)$ and the imaginary part to $\sin(3x)$, we get:

$\cos(3x) = \cos^3(x) - 3\cos(x)\sin^2(x)$

$\sin(3x) = 3\cos^2(x)\sin(x) - \sin^3(x)$

The correct expressions for $\sin(3x)$ are options a:

a. $\sin(3x) = 3\sin(x)\cos^{2}(x) - \sin^{3}(x)$

d. $\sin(3x) = 3\sin(x) - 4\sin^{3}(x)$
