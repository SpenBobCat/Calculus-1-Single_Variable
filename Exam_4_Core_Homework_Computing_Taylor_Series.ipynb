## 3. Compute the Taylor series of $e^{1-\cos x}$ up to degree four.

First, let's find the derivatives:

$y = e^{1-\cos(x)}$

$y' = \frac{d}{dx}(e^{1-\cos(x)}) = e^{1-\cos(x)}\frac{d}{dx}(1-\cos(x)) = e^{1-\cos(x)}\sin(x)$

$y'' = \frac{d^2}{dx^2}(e^{1-\cos(x)}) = \frac{d}{dx}(e^{1-\cos(x)}\sin(x)) = e^{1-\cos(x)}\sin^2(x) + e^{1-\cos(x)}\cos(x)$

$y''' = \frac{d^3}{dx^3}(e^{1-\cos(x)}) = \frac{d}{dx}(e^{1-\cos(x)}\sin^2(x) + e^{1-\cos(x)}\cos(x))$
$= e^{1-\cos(x)}\sin^3(x) + 2e^{1-\cos(x)}\cos(x)\sin(x) - e^{1-\cos(x)}\sin(x)$

$y^{(4)} = \frac{d^4}{dx^4}(e^{1-\cos(x)}) = \frac{d}{dx}(e^{1-\cos(x)}\sin^3(x) + 2e^{1-\cos(x)}\cos(x)\sin(x) - e^{1-\cos(x)}\sin(x))$
$= e^{1-\cos(x)}\sin^4(x) + 3e^{1-\cos(x)}\cos(x)\sin^2(x) + 3e^{1-\cos(x)}\cos^2(x)\sin(x) - e^{1-\cos(x)}\cos(x)$

Now, let's evaluate these derivatives at $x=0$:

$y(0) = e^{1-\cos(0)} = e^1 = e$

$y'(0) = e^{1-\cos(0)}\sin(0) = 0$

$y''(0) = e^{1-\cos(0)}\sin^2(0) + e^{1-\cos(0)}\cos(0) = 1$

$y'''(0) = e^{1-\cos(0)}\sin^3(0) + 2e^{1-\cos(0)}\cos(0)\sin(0) - e^{1-\cos(0)}\sin(0) = 0$

$y^{(4)}(0) = e^{1-\cos(0)}\sin^4(0) + 3e^{1-\cos(0)}\cos(0)\sin^2(0) + 3e^{1-\cos(0)}\cos^2(0)\sin(0) - e^{1-\cos(0)}\cos(0) = \frac{1}{12}$

$e^{1-\cos(x)} = e + 0x + \frac{1}{2!}x^2 + 0x^3 + \frac{1}{4!}x^4 + H.O.T$

Simplifying further, we get:

$e^{1-\cos(x)} = 1 + \frac{1}{2!}x^2 + \frac{1}{12}x^4 + H.O.T$

Therefore, the correct Taylor series of $e^{1-\cos(x)}$ up to degree four is:

$e^{1-\cos(x)} = 1 + \frac{1}{2}x^2 + \frac{1}{12}x^4 + H.O.T$
